##Introduction

The purpose of this project is to analyze bicycle rental data to see when people rent bicycles the most and why. The data set provides information about the weather, time of day, and whether the day was a holiday or not. We can see from these observations the effects of temperature, wind, and even the time of year on the behavior of bicycle renting. This sort of analysis is helpful to the company because it can tell them when to expect more riders and how to plan accordingly. In general, the goal is to find simple patterns in the data for how and when people are using the bikes.

```python
#import the excel
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#load excell file from local system with variable name as "br"
br = pd.read_excel('C:/Windows/System32/bikes_data.xlsx')

br.head(1388)
```

```python
# Print the total number of rows and columns in the dataset, and show basic info about each column
print("Dataset shape (rows, columns):", br.shape)
print("\nColumn information:")
print(br.info())
```

```python
#checking for any missed values from the data set 
print("Missing values in each column:")
print(br.isnull().sum())
```

##Insight 1 

This line graph shows clear spikes in rentals at 8 AM and 5 PM to 6 PM, which are the typical commute hours. These represent that the majority of users rent bikes for commuting to school or work. There are the fewest rentals during early morning and late evenings, which is understandable since travel is less. Midday experiences a smaller but noticeable spike, which can be explained by lunch breaks or recreational rides. Overall, the rental usage is dictated most importantly by time of day, which implies regular utilization rather than events.

```python
#Rental trends for each hour through out the day 

avg_hourly = br.groupby('Hour')['RENTALS'].mean()
avg_hourly.plot(kind='line', marker='o')
plt.title("Hourly Rental Trends Throughout the Day")
plt.xlabel("Hour of Day")
plt.ylabel("Average Rentals")

#adding grid lines
plt.grid(True)
plt.xticks(range(0, 24))

#Displaying the chart
plt.show()
```

##Insight 2


This is a scatter plot showing how temperature affects the amount of bike rentals. It's clear that the most rentals take place when the temperature is between around 15 C and 25 C. This makes sense since people will cycle if the weather isn't too cold or too hot. When the temperature falls below 0 C or rises above 30 C, the rentals fall substantially. The trend suggests pleasant weather has a significant influence on cycle hire demand, and the business can plan more successfully by observing predicted temperatures.

```python
# Plot temperature against number of rentals
plt.scatter(br['Temperature'], br['RENTALS'])
plt.title("Rentals vs Temperature")
plt.xlabel("Temperature (C)")
plt.ylabel("Bike Rentals")
plt.grid(True)

#Display the Scatter plot 
plt.show()
```

##Insight 3 

This bar chart compares the average number of bicycle rentals on non-holidays and holidays. As can be seen, rentals are higher on non-holidays. This likely reflects people renting bicycles for daily transportation to school or work, which is less common on holidays. While there is still some activity on holidays, it is significantly less, reflecting fewer excursions in general. This information can be used by businesses to reduce the number of bikes during holidays or offer rewards for additional riders on holidays.

```python
# Comparing rentals on holidays and non holidays
holiday_avg = br.groupby('Holiday')['RENTALS'].mean()

# Creating of a bar chart
holiday_avg.plot(kind='bar')

#Adding chart titles and labels 
plt.title("Average Rentals: Holiday vs Non-Holiday")
plt.xlabel("Holiday (No/Yes)")
plt.ylabel("Average Rentals")
plt.grid(True)
plt.show()
```

Insight 4

This function categorizes weather as "Good" or "Bad" for cycling rentals based on two simple criteria: temperature and rain. It assumes good weather as between 15 C and 25 C with no rain, which is the traditional wisdom of nice cycling weather. If this condition isn't met, the function qualifies the weather as "Bad." The code then applies this logic to every row in the dataset and prints the result. This binary classification is useful for examining how often perfect biking weather actually occurs, and it can drive planning efforts like dynamic pricing or target-based promotions.

```python
# Create a function to classify weather as good or bad for biking
def rental_weather(temp, rain):
    if temp >= 15 and temp <= 25 and rain == 0:
        return "Good"
    else:
        return "Bad"

# Apply to all the rows
for i in range(1386):
    temp = br.loc[i, 'Temperature']
    rain = br.loc[i, 'Rainfall']
    result = rental_weather(temp, rain)
    print("Row", i, ":", result)
```
